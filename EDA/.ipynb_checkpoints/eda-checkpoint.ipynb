```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
datos=pd.read_csv('datos_eda.csv')
datos.head()
```

```python
#Algunas estadisticas resumidas
datos.describe()
```

```python
#La informacion del dataframe
datos.info()
```

```python
#Convertimos los datos tipo object a un tipo mas adecuado
datos.ubicacion=datos.ubicacion.astype('string')
datos.cole_caracter=datos.cole_caracter.astype('string')
datos.cole_calendario=datos.cole_calendario.astype('string')
datos.cole_depto_ubicacion=datos.cole_depto_ubicacion.astype('string')
datos.cole_jornada=datos.cole_jornada.astype('string')
datos.cole_naturaleza=datos.cole_naturaleza.astype('string')
datos.fami_educacionpadre=datos.fami_educacionpadre.astype('string')
datos.fami_educacionmadre=datos.fami_educacionmadre.astype('string')
datos.fami_estratovivienda=datos.fami_estratovivienda.astype('string')
datos.fami_estratovivienda=datos.fami_estratovivienda.astype('string')
def convertir_si_no(x):
    if x=='Si':
        return True
    elif x=='No':
        return False
    else:
        return np.nan
datos.fami_tieneinternet=datos.fami_tieneinternet.apply(convertir_si_no)
datos.fami_tieneautomovil=datos.fami_tieneautomovil.apply(convertir_si_no)
datos.fami_tienecomputador=datos.fami_tienecomputador.apply(convertir_si_no)
    
    
```

```python
datos.info()
```

```python
datos=datos.dropna()
```

```python
datos.info()
```

```python
datos.fami_tienecomputador=datos.fami_tienecomputador.astype('bool')
datos.fami_tieneinternet=datos.fami_tieneinternet.astype('bool')
datos.fami_tieneautomovil=datos.fami_tieneautomovil.astype('bool')
datos.info()
```

```python
datos.cole_genero=datos.cole_genero.astype('string')
datos.info()
```

```python
my_dict={'Ocho':8,'Siete':7,'Seis':6,'Cuatro':4,'Cinco':5,'Tres':3,'Nueve':9,'3 a 4':4,'Dos':2,'5 a 6':6,'7 a 8':8,'Diez':10,'1 a 2':2,'9 o más':9,'Doce o más':12,'Una':1,'Once':11}
```

```python
datos.fami_personashogar=datos.fami_personashogar.apply(lambda x:my_dict[x])
```

```python
datos.fami_personashogar=datos.fami_personashogar.astype('int')
datos.info()
```

```python
datos.fami_cuartoshogar.unique()
```

```python
my_dict['Seis o mas']=6
my_dict['Diez o más']=10
my_dict['Uno']=1
datos.fami_cuartoshogar=datos.fami_cuartoshogar.apply(lambda x: my_dict[x])
datos.info()
```

```python
def convertir_biling(x):
    if x=='N':
        return False
    else:
        return True
datos.cole_bilingue=datos.cole_bilingue.apply(convertir_biling)
datos.head()
```

```python
datos.cole_bilingue=datos.cole_bilingue.astype('bool')
datos.info()
```

```python
#Iniciamos viendo la influencia de la ubicacion por departamentos
sns.catplot(data=datos, x="ubicacion", y="punt_global", kind="box")
plt.show()
```

```python
sns.catplot(data=datos, x="cole_bilingue", y="punt_global", kind="box")
plt.xlabel('Colegio bilingüe')
plt.ylabel('Puntaje global')
plt.show()
```

```python
index_sort = datos.mean().sort_values().index
index_sort
```

```python
datos.cole_depto_ubicacion=datos.cole_depto_ubicacion.replace('BOGOTÁ','BOGOTA')
```

```python
datos_sem=datos[['cole_depto_ubicacion','punt_global']]
datos_sem=datos_sem.groupby('cole_depto_ubicacion').agg(np.mean)
datos_sem=datos_sem.sort_values('punt_global')
datos_sem.index
```

```python

sns.catplot(data=datos, x="punt_global", y="cole_depto_ubicacion", kind="box",height=8.27, aspect=8.27/8.27,orient="h",order=list(datos_sem.index))
plt.xlabel('puntaje global')
plt.ylabel('Departamento')
plt.show()
```
